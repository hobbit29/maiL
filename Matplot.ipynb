```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
np.arange(20,400,20)
np.arange(0.5,5,0.5)
```

```python
x=np.arange(20, 40, 2)
y=np.arange(10, 40, 3)
plt.title("line graph using matrix data")
plt.plot(x, color= "Green")
plt.xlabel("x-axis")
plt.plot(y, color="red")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.scatter(x, y, color='red', marker="*")
#plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("stock price prediction")
```
