# Q. Implement Simple Linear Regression using analytical method and depict model on scatter data plot. a) Take x=[1,2,4], y=[2,3,6] b) Regress Sales~Radio from Advertisingdata.csv


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def coef(x,y):
    n=np.size(x)
    #mean values of x and y
    mx=np.mean(x)
    my=np.mean(y)
    
    #value of B1 and B0
    B1=(np.sum(x*y)-n*mx*my)/(np.sum(x*x)-n*mx*mx)
    B0=my-B1*mx
    
    return B0,B1

def plotline(x,y,b0,b1):
    plt.scatter(x,y,color="green",marker='o',s=30)
    #response vector
    ypred=b0+b1*x
    
    #plotting regression line
    plt.plot(x,ypred,color="red")
    plt.xlabel('x')
    plt.ylabel('y')
    
x=np.array([1,2,4])
y=np.array([2,3,6])
b0,b1=coef(x,y)
print("Estimated coefficients:\nB0=",b0,"\nB1=",b1)
plotline(x,y,b0,b1)
```

```python
data=pd.read_csv('C:/Users/Arupriya/Downloads/archive (1)/Advertising.csv')
x=data['Radio']
y=data['Sales']

#calculating regression line
def regline(x,y):
    n=len(x)
    mx=np.mean(x)
    my=np.mean(y)
    #coefficients
    b1=(np.sum(x*y)-n*mx*my)/(np.sum(x*x)-n*mx*mx)
    b0=my-b1*mx
    
    plt.scatter(x,y,color="red",marker='o',s=30)
    #response vector
    ypred=b0+b1*x
    
    #plotting regression line
    plt.plot(x,ypred,color="green")
    plt.xlabel('x')
    plt.ylabel('y')
regline(x,y)
```
